## Task 4

With the help of ``glob``, use ``rename_my_file()`` to rename all the ``.txt`` files in the folder.

Note that you might have to adjust your function ``rename_my_file()`` to accommodate the already renamed files.

```python
import glob
import os

# Adjusted rename_my_file function to handle already renamed files
def rename_my_file(old_file_name):
    # Split the file name into components
    file_name, file_ext = os.path.splitext(old_file_name)
    parts = file_name.split('_')
    if len(parts) != 2:
        return old_file_name  # Return the original filename if it's not in the expected format
    
    date, city = parts
    month, day = date.split('-')
    
    # Convert month to numerical format
    months_to_numbers = {
        'jan': '01', 'feb': '02', 'mar': '03',
        'apr': '04', 'may': '05', 'jun': '06',
        'jul': '07', 'aug': '08', 'sep': '09',
        'oct': '10', 'nov': '11', 'dec': '12'
    }
    if month.lower() in months_to_numbers:
        month = months_to_numbers[month.lower()]
    
    # Construct the new file name
    new_file_name = f"{month}-{day}_{city}{file_ext}"
    return new_file_name

# Get a list of all .txt files in the folder
txt_files = glob.glob(r'C:\Users\teren\OneDrive\Desktop\learning-portfolio-Terencetjc\files, folders & os  exercises\os-collaboration-exercise-data\*.txt')

# Rename each .txt file using rename_my_file function
for txt_file in txt_files:
    new_file_name = rename_my_file(os.path.basename(txt_file))
    new_file_path = os.path.join(os.path.dirname(txt_file), new_file_name)
    os.rename(txt_file, new_file_path)
    print(f"{txt_file} ----> {new_file_name}")
```

```python
import os
import glob
import shutil

# Define the file path folder
file_path_folder = r'C:\Users\teren\OneDrive\Desktop\learning-portfolio-Terencetjc\files, folders & os  exercises\os-collaboration-exercise-data'

# Define the folder for Cairo
cairo_folder = os.path.join(file_path_folder, 'Cairo')

# Use glob to list all the files from the file path folder
all_files = glob.glob(os.path.join(file_path_folder, '*.txt'))

# Filter the files related to Cairo
cairo_files = [file for file in all_files if 'cairo' in file.lower()]

# Copy files related to Cairo to the Cairo folder
for file in cairo_files:
    shutil.copy(file, cairo_folder)
    print(f"File {file} copied to {cairo_folder}")
```

```python
import os
import glob
import shutil

# Define the file path folder
file_path_folder = r'C:\Users\teren\OneDrive\Desktop\learning-portfolio-Terencetjc\files, folders & os  exercises\os-collaboration-exercise-data'

# Define the list of cities
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

# Iterate over each city
for city in cities:
    # Define the folder for the current city
    city_folder = os.path.join(file_path_folder, city)
    
    # Use glob to list all the files from the file path folder
    all_files = glob.glob(os.path.join(file_path_folder, '*.txt'))

    # Filter the files related to the current city
    city_files = [file for file in all_files if city.lower() in file.lower()]

    # Copy files related to the current city to the city folder
    for file in city_files:
        shutil.copy(file, city_folder)
        print(f"File {file} copied to {city_folder}")
```
