# A simple plot

## Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt  # Importing the pyplot module from matplotlib and aliasing it as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]  # X-values for the data points
y_1 = [0, 2, 4, 6, 8, 10, 12]  # Y-values for the first set of data points
y_2 = [0, 3, 6, 9, 12, 15, 18]  # Y-values for the second set of data points

# Let's start plotting
plt.plot(x, y_1, label='Y values', color='red', linestyle='dashed')  # Plotting the first set of data points with a red dashed line
plt.xlabel('x-values')  # Labeling the x-axis
plt.ylabel('y-values')  # Labeling the y-axis
plt.title('X vs Y')  # Setting the title of the plot
plt.grid(alpha=0.8)  # Adding a grid to the plot with 80% transparency
plt.legend(loc='upper left')  # Adding a legend to the plot in the bottom-left corner
```

```python
import numpy as np
import scipy
from scipy.stats import linregress
from matplotlib import pyplot as plt

# Some data for plotting
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([0, 2, 4, 6, 8, 10, 12])

# Fit a linear regression line to the data
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Derive the equation of the regression line
equation = f'y = {slope:.2f}x + {intercept:.2f}'

# Plot the data points and the regression line
plt.plot(x, y, 'o', label='Data points')
plt.plot(x, slope * x + intercept, 'r-', label='Regression line')

# Annotate the equation on the graph
plt.text(3, 9, equation, fontsize=12)

# Add labels, title, grid, and legend
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y with Regression Line')
plt.grid(alpha=0.8)
plt.legend()

# Show the plot
plt.show()
```

## Yet another plot but with error bars

```python
#Include x and y error bars for the points using errorbar(). 
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.errorbar(x, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green")
```

##  Adding mathematical functions

```python
# Note: NumPy arrays allow us to generate data-related mathematical functions.
# Example here which plots x^2 and sin(x).
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

## We need another axis!

```python
#Have another y-axis that shares the same x-axis using command twinx().
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Create an array of values from 0 to 6 using np.linspace()
x = np.linspace(0, 6, 100)  # Generates 100 evenly spaced values between 0 and 6

# Calculate the square of each value in x
x2 = x**2

# Calculate the sine of each value in x
sin_x = np.sin(x)

# Plot x vs. x^2 with a red dashed line, and label it as 'x^2'
plt.plot(x, x2, label='x^2', color='red', linestyle='dashed')

# Add a legend to the plot and position it in the lower left corner
plt.legend(loc='lower left')

# Create a new y-axis for the plots that come after
plt.twinx()

# Plot x vs. sin(x) with a blue line, and label it as 'sin(x)'
plt.plot(x, sin_x, label='sin(x)', color='blue')

# Add a legend to the plot and position it in the lower right corner
plt.legend(loc='lower right')
```

# A real example: Global Warming

## Plotting data from files

Plotting data stored in a file (e.g., spreadsheet, text file, database) is a routine task for a scientist. In fact, the first thing you should do with any data is to look at it with a simple plot.

For the rest of this section, I will use the Earthâ€™s land temperature data from the [Berkeley Earth website](https://berkeleyearth.org/). Please visit the site (Global Warming --> Data Overview) and download the average temperature data for Daily Land. The original name of the file should be ``Complete_TAVG_daily.txt``

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5) #used a small alpha value to soften the colour of the plot and made the plot range symmetrical in the y direction.
plt.ylim([-8, 8])

#hlines() function needs a y-value and starting and ending values for x. 
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') #add a horizontal line at the zero value to highlight the trend shown by the data.
```

## Styles

Most people want a decent-looking plot without spending time customising it. To facilitate this Matplotlib offers some standard style templates (see [here](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html)). I am going to use the one called **fivethirtyeight**.

```python
import numpy as np
# Set the plot style to 'fivethirtyeight'
plt.style.use('fivethirtyeight')

# Load data from the file 'Complete_TAVG_daily.txt', skipping the first 24 rows
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

# Extract the date column from the data
date = data[:, 0]

# Extract the temperature anomaly column from the data
anomaly = data[:, -1]

# Plot the date against the temperature anomaly with transparency set to 0.5
plt.plot(date, anomaly, alpha=.5)

# Add a horizontal dashed line at y=0 to indicate the average temperature
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')

# Set the y-axis limits to -8 and 8
plt.ylim([-8, 8])

# Label the x-axis as 'Date'
plt.xlabel('Date')

# Label the y-axis as 'Temperature Anomaly'
plt.ylabel('Temperature Anomaly')

# Set the title of the plot as 'Temperature anomaly (Relative to average from Jan 1951 - Dec 1980.)'
plt.title('Temperature anomaly\n(Relative to average from Jan 1951 - Dec 1980.)')
```

### xkcd!

Since we are talking about styles, I must tell you that the developers of Matplotlib have a healthy sense of humour and have included the option of making your plots in the xkcd style. To enable this, just run ``plt.xkcd()`` instead of setting a style. Here is what the previous plot looks like using the xkcd style.

```python
from matplotlib import pyplot as plt
import numpy as np
plt.xkcd()
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

#This style uses specific font family 'xkcd', 'xkcd script' and 'Comic Neue' but I'm not going to manually download and use fonts so ...
# Specify a different font family
plt.rcParams['font.family'] = 'Arial'  # You can replace the style font with any font available in the system eg. 'Arial' 

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
