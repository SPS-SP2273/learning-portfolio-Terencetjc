# Exercise 1

```python
#Fill in the blanks(indicated with a ?), so the code below will produce the plot shown
from matplotlib import pyplot as plt
import numpy as np
# Some data for plotting
x = [0, 1, 2, 3, 4, 5] #x-values
y = [0, 4, 8, 12, 16, 20] #y-values
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5] # Error in y-values
x_err = .25 # Error in x-values (Constant for all points)

#Dark red color(RGB:139, 0, 0)
dark_red=(139/255, 0, 0)

#Create error plot using errorbar function
# x: x-values for the data points
# y: y-values for the data points
# xerr: Error in x-values (optional, default=None)
# yerr: Error in y-values (optional, default=None)
# color: Color of the line (optional, default=None)
# linestyle: Line style of the error bars (optional, default='-')
# marker: Marker style for the data points (optional, default=None)
# markerfacecolor: Face color of the markers (optional, default=None)
# ecolor: Color of the error bars (optional, default=None)
# capsize: Length of the error bar caps (optional, default=None)
# label: Label for the legend (optional, default=None)
plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color=dark_red, linestyle='--',
             marker='o', markerfacecolor="yellow",
             ecolor="black",capsize=5,
             label="Y values with errors")
# Set label for the x-axis
plt.xlabel('x-values')
# Set label for the y-axis
plt.ylabel('y-values')
# Set title for the plot
plt.title('X vs Y')
# Add gridlines to the plot with transparency (alpha)
plt.grid(alpha=.25)
# Add a legend to the plot
plt.legend(loc='lower right')
# Display the plot
plt.show()
```

# Exercise 2 (A simple mathematical plot) 

1. Use NumPy to generate x and y data suitable for plotting the graph of the function:

y=$e^{(-x/10)}sin(x)$

for values of x in the range 0 to 50.

3. Customise your plot so that:
- The x and y axes have labels of ``fontsize`` of 15
- The x and y axes labels are in the ``color``, ``darkorange``.
- There is a grid with an opacity of 25%.
- Indicate the equation of the function in the title with a font size of 20. <br>
Hint: Look at [Writing mathematical expressions](https://matplotlib.org/stable/users/explain/text/mathtext.html) in Matplotlib.
- Change the y limit (using ``ylim``) to [-1,1]
In the end, your plot should look like the one shown alongside. <br>

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate x values
x = np.linspace(0, 50, 500)

# Step 2: Calculate y values
y = np.exp(-x / 10) * np.sin(x)

# Step 3: Create the plot
plt.plot(x, y)

# Step 4: Customize the plot
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange')
plt.title(r'$y = e^{-\frac{x}{10}} \sin(x)$', fontsize=20)
plt.grid(alpha=0.25)
plt.ylim([-1, 1])

# Display the plot
plt.show()
```

```python
#Example for experiments data point?
import numpy as np  # Import NumPy library for numerical computations
import matplotlib.pyplot as plt  # Import matplotlib.pyplot for plotting

# Generate x and y data
x = np.linspace(0, 50, 500)  # Create an array of 500 evenly spaced values from 0 to 50 for x-axis
y = np.exp(-x / 10) * np.sin(x)  # Calculate y-values using the given mathematical expression

# Add some random uncertainties to y
y_err = np.random.uniform(0.1, 0.5, size=len(x))  # Generate random uncertainties for each y-value

# Customize plot
plt.figure(figsize=(8, 6))  # Set the figure size to 8 inches by 6 inches

# Plot with error bars
plt.errorbar(x, y, yerr=y_err, fmt='o', markersize=3, label='Data with Uncertainty')  # Plot data points with error bars

# Fit a curve
p = np.polyfit(x, y, 3)  # Fit a polynomial curve of degree 3 to the data points
plt.plot(x, np.polyval(p, x), 'r--', label='Fit Curve')  # Plot the fitted curve

# Customize legend
plt.legend(loc='upper right', fontsize=12)  # Display legend at upper right with fontsize 12

# Add title and labels
plt.title('Plot of $e^{-x/10} \sin(x)$ with Error Bars and Fit Curve', fontsize=20)  # Set the title of the plot
plt.xlabel('x', fontsize=15, color='darkorange')  # Set the label for the x-axis with fontsize 15 and dark orange color
plt.ylabel('y', fontsize=15, color='darkorange')  # Set the label for the y-axis with fontsize 15 and dark orange color

# Add grid with opacity
plt.grid(alpha=0.25)  # Add gridlines with opacity of 25%

# Change y limit
plt.ylim(-1, 1)  # Set the y-axis limit from -1 to 1

# Save the plot as an image
plt.savefig('plot.png', dpi=300)  # Save the plot as an image with a resolution of 300 dpi

# Show plot
plt.show()  # Display the plot
```
